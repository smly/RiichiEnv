# RiichiEnv Replay Viewer Demo

This notebook demonstrates how to use the standalone Replay Viewer using a sample log provided in the repo.  
The viewer works completely client-side in the browser, meaning you can save this notebook as HTML or view it on NBViewer and the replay will still be interactive.

```python
import riichienv
from riichienv.visualizer import show_replay
import json
```

```python
# Load the sample log
log_path = '../tools/replay-visualizer/example_log_b124_m244_0.jsonl'

with open(log_path, 'r') as f:
    events = [json.loads(line.strip()) for line in f]

print(f"Loaded {len(events)} events.")
```

```python
# Display the Replay Viewer
# This embeds the JS and Assets directly into the output cell.
show_replay(events)
```

### Notes
- Controls: Use **Prev** and **Next** to step through the log, or drag the **Slider**.
- Logic: The viewer replays the events in JavaScript to reconstruct the state at each step.
