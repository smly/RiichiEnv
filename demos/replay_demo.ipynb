# RiichiEnv Replay Viewer Demo

This notebook demonstrates how to use the standalone Replay Viewer using a sample log provided in the repo.  
The viewer works completely client-side in the browser, meaning you can save this notebook as HTML or view it on NBViewer and the replay will still be interactive.

```python
from riichienv import RiichiEnv
from riichienv.agents import RandomAgent


agent = RandomAgent()
env = RiichiEnv()
obs_dict = env.reset()
while not env.done():
    actions = {player_id: agent.act(obs)
               for player_id, obs in obs_dict.items()}
    obs_dict = env.step(actions)

scores, points, ranks = env.scores(), env.points(), env.ranks()
print(scores, points, ranks)
```

```python
from riichienv.visualizer import Replay

Replay.from_list(env.mjai_log)
```
